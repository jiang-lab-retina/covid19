```python
import csv
file = csv.DictReader(open("overwatch.csv"))
pokemon_list = []
for row in file:
    print(dict(row))
    pokemon_list.append(dict(row))
    
```

```python
for pokemon in pokemon_list:
    for key in pokemon:
        try:
            pokemon[key] = int(pokemon[key])
        except:
            pass
    if pokemon["Legendary"]== "True":
        pokemon["Legendary"]=True
    else:
        pokemon["Legendary"]= False
```
